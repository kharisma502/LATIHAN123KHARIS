```python
# Operasi Matematika di Python (Latihan 0)

# Penjumlahan
print(2 + 3 + 9)  # Output: 14

# Pengurangan
print(99 - 73)  # Output: 26

# Perkalian
print(23.54 * -1432)  # Output: -33709.28

# Pembagian (hasil float)
print(100 / 7)  # Output: 14.285714285714286

# Pembagian (hasil integer)
print(100 // 7)  # Output: 14

# Modulus (sisa pembagian)
print(100 % 7)  # Output: 2

# Pangkat
print(5 ** 3)  # Output: 125
```

```python
#Latihan 1

# Variabel
my_favorite_number = 1
my_least_favorite_number = 5
a_neutral_number = 3

# Pengecekan Kesetaraan (Equality Check) - True
print(my_favorite_number == 1)  # Output: True

# Pengecekan Kesetaraan (Equality Check) - False
print(my_favorite_number == my_least_favorite_number)  # Output: False

# Pengecekan Ketidaksamaan (Not Equal Check) - True
print(my_favorite_number != a_neutral_number)  # Output: True

# Pengecekan Ketidaksamaan (Not Equal Check) - False
print(a_neutral_number != 3)  # Output: False

# Pengecekan Lebih Besar (Greater Than) - True
print(my_least_favorite_number > a_neutral_number)  # Output: True
```

```python
# Import library
import sklearn
from sklearn import datasets
from sklearn import tree
from sklearn.model_selection import cross_val_score

# Load iris dataset
iris = datasets.load_iris()

# Mendefinisikan atribut dan label pada dataset
X = iris.data
y = iris.target

# Membuat model dengan Decision Tree Classifier
clf = tree.DecisionTreeClassifier()

# Mengevaluasi performa model dengan cross_val_score
scores = cross_val_score(clf, X, y, cv=5)

# Menampilkan hasil cross-validation
print("Akurasi pada setiap fold:", scores)
print("Rata-rata akurasi:", scores.mean())
```

```python
# Import library
from sklearn.datasets import load_iris

# Load dataset iris dengan X (fitur) dan y (label)
X, y = load_iris(return_X_y=True)

# Menampilkan dimensi fitur dan kelas pada dataset
print(f"Dimensi Feature: {X.shape}")
print(f"Class: {set(y)}")

# Alternatif: Memuat dataset sebagai objek
data = load_iris()
print(data)
```

```python
# Import library untuk membagi dataset menjadi training dan testing
from sklearn.model_selection import train_test_split

# Membagi dataset menjadi data training dan testing (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=0)

# Import DecisionTreeClassifier dari sklearn
from sklearn.tree import DecisionTreeClassifier

# Membuat model Decision Tree dengan kedalaman maksimum 4
model = DecisionTreeClassifier(max_depth=4)

# Melatih model dengan data training
model.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

# Mengatur ukuran gambar
plt.subplots(figsize=(10,10))

# Menampilkan Decision Tree
tree.plot_tree(model, fontsize=10)
plt.show()
```

```python
#eval model

from sklearn.metrics import classification_report

# Melakukan prediksi pada data uji
y_pred = model.predict(X_test)

# Menampilkan laporan klasifikasi
print(classification_report(y_test, y_pred))
```

```python
# Data jumlah kamar
bedrooms = np.array([1,1,2,2,3,4,4,5,5,5])

# Data harga rumah
house_price = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000, 85000, 90000])
```

```python
from sklearn.linear_model import LinearRegression

# Reshape agar bisa diproses oleh LinearRegression
bedrooms = bedrooms.reshape(-1,1)

# Membuat model regresi linear
linreg = LinearRegression()
linreg.fit(bedrooms, house_price)
```

```python
plt.scatter(bedrooms, house_price, label="Data Asli")
plt.plot(bedrooms, linreg.predict(bedrooms), color='red', label="Regresi Linear")
plt.xlabel("Jumlah Kamar")
plt.ylabel("Harga Rumah")
plt.title("Regresi Linear: Jumlah Kamar vs Harga Rumah")
plt.legend()
plt.show()
```

```python
# Gantilah path dataset sesuai lokasi file kamu
df = pd.read_csv('/content/Mall_Customers.csv')

# Menampilkan 5 data pertama
df.head()
```

```python
# Mengubah nama kolom
df = df.rename(columns={
    'Gender': 'gender',
    'Age': 'age',
    'Annual Income (k$)': 'annual_income',
    'Spending Score (1-100)': 'spending_score'
})

# Mengubah data kategorik menjadi data numerik
df['gender'].replace(['Female', 'Male'], [0, 1], inplace=True)

# Menampilkan 5 data pertama
df.head()
```

```python
# Delete Kolom CustomerID dan gender
X = df.drop(['CustomerID', 'gender'], axis=1)
```

```python
# Import Algoritma
from sklearn.cluster import KMeans
```

```python
# Membuat List dari nilai K 1-11 (inertia_)
from sklearn.cluster import KMeans

clusters = []
for i in range(1, 11):
    km = KMeans(n_clusters=i)  # Perbaikan nama variabel
    km.fit(X)  # Perbaikan sintaks `.fit(X)`
    clusters.append(km.inertia_)  # Memastikan `km.inertia_` ditambahkan ke list dengan benar
```

```python
# import Library untuk visualisasi
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat plot inertia
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)

ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')  # Perbaikan tanda kutip
plt.show()  # Menampilkan plot
```

```python
from sklearn.cluster import KMeans

# Membuat objek KMeans
km5 = KMeans(n_clusters=5, random_state=42).fit(X)

# Menambahkan kolom Label pada dataset
X['Labels'] = km5.labels_
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(8, 4))
sns.scatterplot(x=X['annual_income'], y=X['spending_score'], hue=X['Labels'], palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.show()
```
